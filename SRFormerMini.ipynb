```python
import os
import shutil

# SRFormer klasörünün yolu
src_folder = "./SRFormer"

# SRFormer'in içeriğinin taşınacağı hedef klasör
dst_folder = "./"  # Aynı dizinde dışarı taşımak için "." kullanılır

# SRFormer klasörünün içindeki her şeyi taşı
for item in os.listdir(src_folder):
    src_path = os.path.join(src_folder, item)
    dst_path = os.path.join(dst_folder, item)

    try:
        if os.path.isfile(src_path) or os.path.islink(src_path):
            shutil.move(src_path, dst_path)  # Dosyayı taşı
        elif os.path.isdir(src_path):
            shutil.move(src_path, dst_path)  # Alt klasörü taşı
        print(f"Taşındı: {src_path} -> {dst_path}")
    except Exception as e:
        print(f"Taşınamadı: {src_path} -> {dst_path} (Hata: {e})")

# SRFormer klasörünü sil (boş kaldıysa)
if not os.listdir(src_folder):
    os.rmdir(src_folder)
    print(f"SRFormer klasörü silindi.")
else:
    print(f"SRFormer klasörü boşaltılamadı.")
```

```python
import os
import zipfile

def extract_zip_to_folder(zip_path, output_folder):
    """
    Verilen zip dosyasını belirtilen hedef klasöre çıkarır.

    Args:
        zip_path (str): Çıkarılacak zip dosyasının tam yolu.
        output_folder (str): Dosyaların çıkarılacağı hedef klasör.
    """
    # Hedef klasörü oluştur (varsa hata vermez)
    os.makedirs(output_folder, exist_ok=True)

    # Zip dosyasını çıkar
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(output_folder)

    print(f"Files have been successfully extracted from {zip_path} to {output_folder}.")


output_folder_path = "/content/datasets"

# Zip dosyasını hedef klasöre çıkar
extract_zip_to_folder("/content/drive/MyDrive/dataset/Set5.zip", output_folder_path)
extract_zip_to_folder("/content/drive/MyDrive/dataset/Set14.zip", output_folder_path)
extract_zip_to_folder("/content/drive/MyDrive/dataset/BSDS100.zip", output_folder_path)
extract_zip_to_folder("/content/drive/MyDrive/dataset/manga109.zip", output_folder_path)
extract_zip_to_folder("/content/drive/MyDrive/dataset/urban100.zip", output_folder_path)
```

```python
import os
import cv2

def generate_gtmod12_and_lr(hr_folder, output_folder, scales=[2, 3, 4]):
    """
    HR görüntülerden GTmod12 ve LR görüntüleri oluşturur.

    Args:
        hr_folder (str): Orijinal HR görüntülerin bulunduğu klasör.
        output_folder (str): Çıktıların kaydedileceği ana klasör.
        scales (list): LR görüntülerin oluşturulacağı ölçek faktörleri (ör. [2, 3, 4]).
    """
    # GTmod12 klasörünü oluştur
    gtmod12_folder = os.path.join(hr_folder, "GTmod12")
    os.makedirs(gtmod12_folder, exist_ok=True)

    # LR klasörlerini oluştur
    lr_folders = {scale: os.path.join(output_folder, f"x{scale}") for scale in scales}
    for lr_folder in lr_folders.values():
        os.makedirs(lr_folder, exist_ok=True)

    for filename in os.listdir(hr_folder):
        hr_path = os.path.join(hr_folder, filename)

        # Sadece görüntü dosyalarını işle
        if filename.endswith(('.png', '.jpg', '.jpeg')):
            # HR görüntüyü yükle
            img = cv2.imread(hr_path)
            if img is None:
                print(f"Error loading file: {hr_path}")
                continue

            # GTmod12 için kırpma işlemi
            height, width = img.shape[:2]
            height_mod = height - (height % max(scales))
            width_mod = width - (width % max(scales))
            gtmod12_img = img[:height_mod, :width_mod]

            # GTmod12 görüntüyü kaydet
            gtmod12_path = os.path.join(gtmod12_folder, filename)
            cv2.imwrite(gtmod12_path, gtmod12_img)
            print(f"Saved GTmod12 image: {gtmod12_path}")

            # LR görüntüler oluştur ve kaydet
            for scale in scales:
                lr_image = cv2.resize(gtmod12_img,
                                      (width_mod // scale, height_mod // scale),
                                      interpolation=cv2.INTER_CUBIC)
                lr_path = os.path.join(lr_folders[scale], filename)
                cv2.imwrite(lr_path, lr_image)
                print(f"Saved LR x{scale} image: {lr_path}")

```

```python
# HR klasörünün yolu
hr_folder = "/content/datasets/BSDS100"

# LR görüntülerin kaydedileceği ana klasör
output_base_folder = "/content/datasets/BSDS100/LR_bicubic"

# HR görüntülerden LR görüntüleri üret Set5 ve Set14 zaten bu klasörleri içeriyor.
generate_gtmod12_and_lr("/content/datasets/BSDS100", "/content/datasets/BSDS100/LR_bicubic", scales=[2, 3, 4])
generate_gtmod12_and_lr("/content/datasets/urban100", "/content/datasets/urban100/LR_bicubic", scales=[2, 3, 4])
generate_gtmod12_and_lr("/content/datasets/manga109", "/content/datasets/manga109/LR_bicubic", scales=[2, 3, 4])
```

```python
import os
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
from skimage.io import imread
import pandas as pd

def calculate_metrics(gt_folder, sr_folder):
    """
    GT (High Resolution) ve SR (Super Resolution) klasörlerini karşılaştırarak PSNR ve SSIM metriklerini hesaplar.

    Args:
        gt_folder (str): Yüksek çözünürlüklü (GT) görüntülerin bulunduğu klasör yolu.
        sr_folder (str): Süper çözünürlük (SR) görüntülerin bulunduğu klasör yolu.

    Returns:
        pd.DataFrame: PSNR ve SSIM değerlerini içeren bir DataFrame.
    """
    def crop_to_match(hr_img, sr_img):
        """HR ve SR boyutlarını eşleştirir. SR fazladan pikselleri kırpar."""
        H, W = hr_img.shape[:2]
        return sr_img[:H, :W]

    # Sonuçları saklamak için liste
    results = []

    # GT klasöründeki dosyaları dolaş
    for filename in os.listdir(gt_folder):
        file1_path = os.path.join(gt_folder, filename)
        file2_path = os.path.join(sr_folder, filename)

        # Dosyalar mevcutsa işle
        if os.path.isfile(file1_path) and os.path.isfile(file2_path):
            # Görüntüleri yükle
            img1 = imread(file1_path)  # High Resolution (HR) görüntü
            img2 = imread(file2_path)  # Super Resolution (SR) görüntü

            # Görüntü boyutlarını eşitle (fazladan pikselleri kırp)
            img2 = crop_to_match(img1, img2)

            # PSNR ve SSIM hesapla
            score_psnr = psnr(img1, img2, data_range=img1.max() - img1.min())
            score_ssim = ssim(img1, img2, data_range=img1.max() - img1.min(), channel_axis=-1)

            # Sonuçları kaydet
            results.append({
                "Filename": filename,
                "PSNR": score_psnr,
                "SSIM": score_ssim
            })

    # Sonuçları DataFrame olarak oluştur
    df = pd.DataFrame(results)

    # Ortalama PSNR ve SSIM hesapla
    avg_psnr = df['PSNR'].mean() if not df.empty else 0
    avg_ssim = df['SSIM'].mean() if not df.empty else 0

    # Tam dosya yolunu yazdır
    print(f"Input Folder: {sr_folder}")
    print(f"Average PSNR: {avg_psnr:.2f}")
    print(f"Average SSIM: {avg_ssim:.4f}")

    return df


```

```python
import os
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
from skimage.io import imread
import pandas as pd

def calculate_metrics(gt_folder, sr_folder):
    """
    GT (High Resolution) ve SR (Super Resolution) klasörlerini karşılaştırarak PSNR ve SSIM metriklerini hesaplar.

    Args:
        gt_folder (str): Yüksek çözünürlüklü (GT) görüntülerin bulunduğu klasör yolu.
        sr_folder (str): Süper çözünürlük (SR) görüntülerin bulunduğu klasör yolu.

    Returns:
        pd.DataFrame: PSNR ve SSIM değerlerini içeren bir DataFrame.
    """
    def crop_to_match(hr_img, sr_img):
        """HR ve SR boyutlarını eşleştirir. SR fazladan pikselleri kırpar."""
        H, W = hr_img.shape[:2]
        return sr_img[:H, :W]

    # Sonuçları saklamak için liste
    results = []

    # GT klasöründeki dosyaları dolaş
    for filename in os.listdir(gt_folder):
        file1_path = os.path.join(gt_folder, filename)
        file2_path = os.path.join(sr_folder, filename)

        # Dosyalar mevcutsa işle
        if os.path.isfile(file1_path) and os.path.isfile(file2_path):
            # Görüntüleri yükle
            img1 = imread(file1_path)  # High Resolution (HR) görüntü
            img2 = imread(file2_path)  # Super Resolution (SR) görüntü

            # Görüntü boyutlarını eşitle (fazladan pikselleri kırp)
            img2 = crop_to_match(img1, img2)

            # PSNR ve SSIM hesapla
            score_psnr = psnr(img1, img2, data_range=img1.max() - img1.min())
            score_ssim = ssim(img1, img2, data_range=img1.max() - img1.min(), channel_axis=-1)

            # Sonuçları kaydet
            results.append({
                "Filename": filename,
                "PSNR": score_psnr,
                "SSIM": score_ssim
            })

    # Sonuçları DataFrame olarak oluştur
    df = pd.DataFrame(results)

    # Ortalama PSNR ve SSIM hesapla
    avg_psnr = df['PSNR'].mean() if not df.empty else 0
    avg_ssim = df['SSIM'].mean() if not df.empty else 0

    # Tam dosya yolunu yazdır
    print(f"Input Folder: {sr_folder}")
    print(f"Average PSNR: {avg_psnr:.2f}")
    print(f"Average SSIM: {avg_ssim:.4f}")

    return df


```

```python
# PSNR ve SSIM metriklerini hesapla
print(calculate_metrics("/content/datasets/BSDS100/GTmod12", "/content/resulting/output/BSDS100/globalx4"))
print(calculate_metrics("/content/datasets/Set14/GTmod12", "/content/resulting/output/Set14/globalx4"))
print(calculate_metrics("/content/datasets/Set5/GTmod12", "/content/resulting/output/Set5/globalx4"))
print(calculate_metrics("/content/datasets/manga109/GTmod12", "/content/resulting/output/manga109/globalx4"))
print(calculate_metrics("/content/datasets/urban100/GTmod12", "/content/resulting/output/urban100/globalx4"))
```
